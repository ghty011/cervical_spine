```python
from PIL import Image
import pandas as pd
import numpy as np
import os
import sys
import glob
import re
from tqdm import tqdm
```

```python
DATA_DIR = "/root/autodl-tmp/cervical_spine/"

MASK_DIR = os.path.join(DATA_DIR, 'segmentation_axial_results_132508')
```

```python
test_slices = glob.glob(f'{MASK_DIR}/*/*')
print(len(test_slices))
test_slices = [re.findall(f'{MASK_DIR}/(.*)/(.*).png', s)[0] for s in test_slices]

df_test_slices = pd.DataFrame(data=test_slices, columns=['StudyInstanceUID', 'Slice'])
df_test_slices['UID_Slice'] = df_test_slices['StudyInstanceUID'] + '.' + df_test_slices['Slice'].astype('string')
df_test_slices = df_test_slices.set_index('UID_Slice').astype({'Slice': int})
df_test_slices
```

```python
s = df_test_slices.iloc[1]
path = os.path.join(MASK_DIR, s.StudyInstanceUID, f'{s.Slice}.png')

mask = Image.open(path)
mask = np.asarray(mask) / 256.


print(mask.shape, mask.min(), mask.max())
mask[mask > 0].mean()
```

```python
cvalues = []
for i in tqdm(range(len(df_test_slices))):
    s = df_test_slices.iloc[i]
    path = os.path.join(MASK_DIR, s.StudyInstanceUID, f'{s.Slice}.png')

    mask = Image.open(path)
    mask = np.asarray(mask) / 256.

    c = mask[mask > 0].mean()
    cvalues.append(c)
```

```python
# cvalues
```

```python
df_test_slices['CV'] = cvalues
```

```python
df_test_slices = df_test_slices.fillna(0)
df_test_slices
```

```python
df_test_slices.to_csv(os.path.join(DATA_DIR, 'cvalues_for_segmentation_axial_results_132508.csv'))
```
